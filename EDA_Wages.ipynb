```python
import pandas as pd
```

```python
df_wage = pd.read_csv('Datasets/SLID.csv')
```

```python
df_wage.head()
```

```python
df_wage.shape
```

```python
df_wage.drop(columns = 'Unnamed: 0', inplace = True)
```

```python
from plotnine import *
```

```python
plt_wage2 = ggplot(df_wage, aes(x ='education', y ='wages'))+ geom_point(aes(color = 'language')) + facet_grid('~sex')
print(plt_wage2)
```

```python
df_wage.head()
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
sns.set()

plt.figure(figsize=(15,10))
plt.subplot(1,3,1)
sns.boxplot(x = 'sex', y = 'wages', data = df_wage)

plt.subplot(1,3,2)
sns.boxplot(x = 'sex', y = 'education', data = df_wage)

plt.subplot(1,3,3)
sns.boxplot(x = 'sex', y = 'age', data = df_wage)
```

```python
plt.figure(figsize=(15,10))
plt.subplot(1,3,1)
sns.distplot(df_wage['wages'])

plt.subplot(1,3,2)
sns.distplot(df_wage['education'])

plt.subplot(1,3,3)
sns.distplot(df_wage['age'])
```

```python
df_wage.describe()
```

```python
plt_dist = ggplot(df_wage, aes(x= 'sex', fill = 'language')) + geom_bar()
print(plt_dist)
```

```python
plt_bubble = ggplot(df_wage, aes(x = 'education', y = 'wages', color = 'language', size = 'age')) + geom_point() + theme_bw()
```

```python
print(plt_bubble)
```

```python
import numpy as np
fig, ax = plt.subplots(figsize= (15,8))

wages_corr = df_wage.corr()
matrix = np.triu(wages_corr)
sns.heatmap(wages_corr, annot = True, mask = matrix)
```

```python
sns.pairplot(df_wage, hue ='sex')
```
