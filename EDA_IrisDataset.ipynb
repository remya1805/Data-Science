### Determine distribution of features: 

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
sns.boxplot(x= 'species', y = 'sepal length (cm)', data = df_iris )

plt.subplot(2,2,2)
sns.boxplot(x= 'species', y = 'sepal width (cm)', data = df_iris )

plt.subplot(2,2,3)
sns.boxplot(x= 'species', y = 'petal length (cm)', data = df_iris )

plt.subplot(2,2,4)
sns.boxplot(x= 'species', y = 'petal width (cm)', data = df_iris )
```

```python
plt.figure(figsize=(12,10))

plt.subplot(2,2,1)
sns.violinplot(x= 'species', y = 'sepal length (cm)', data = df_iris )

plt.subplot(2,2,2)
sns.violinplot(x= 'species', y = 'sepal width (cm)', data = df_iris )

plt.subplot(2,2,3)
sns.violinplot(x= 'species', y = 'petal length (cm)', data = df_iris )

plt.subplot(2,2,4)
sns.violinplot(x= 'species', y = 'petal width (cm)', data = df_iris )
```

Observations:
1. Setosa has smaller sepal length, petal length, petal width compartively but higher sepal width.
2. Not a lot of skew in the data
3. Not many outliers
4. Unimodal

### Determine segmentation

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x = 'sepal length (cm)', y='sepal width (cm)', data=df_iris, marker ='D', hue='species')
```

```python
sns.pairplot(data = df_iris, hue='species')
```

Observations:
1. Petal length and petal width can be best used to distinguish between the three categories since there is least overlap for versicolor and virginica.
2. Sepal Length and Sepal width are not good features to distinguish between the categories since there is a lot of overlap.
