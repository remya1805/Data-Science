## The iris dataset

#### Dataset description
**Iris:** This iris data set gives the measurements in centimeters of the variables sepal length and width and petal length and width, respectively, for 50 flowers from each of 3 species of iris. The species are Iris setosa, versicolor, and virginica. __[Click here to read more](https://www.rdocumentation.org/packages/datasets/versions/3.6.2/topics/iris)__.

```python
# Import the required package
import pandas as pd
# Read and preview the data
iris = pd.read_csv('Datasets/Iris_Data.csv')
print(iris.head())
```

### Box plots
Summarizes quantitative/numeric data.

```python
# Import the required packages
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12,8))
# Seaborn boxplot
sns.boxplot(x='Species', y='Sepal.Length', data=iris)
plt.xlabel("Species", fontsize=16)
plt.ylabel("Sepal Length", fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

### Pie charts
Summarizes qualitative/categorical variables.

```python
# Pandas pie chart
plt.figure(figsize=(12,8))
iris['Species'].value_counts().plot(kind='pie')
plt.show()
```

### Scatter plots
Visual depiction of correlation between numeric variables.

```python
# Seaborn scatter plot
plt.figure(figsize=(12,8))
sns.scatterplot(x=iris['Sepal.Length'], y=iris['Sepal.Width'])
plt.xlabel("Sepal Length", fontsize=16)
plt.ylabel("Sepal Width", fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

## The diamonds dataset

#### Dataset description
**Diamonds:** A dataset containing the prices and other attributes of almost 54,000 diamonds. __[Click here to read more](https://ggplot2.tidyverse.org/reference/diamonds.html)__.

```python
# Read and preview the data
diamonds = pd.read_csv('Datasets/diamonds.csv')
print(diamonds.head())
```

### ggplot fundamentals
- `ggplot()` provides a blank canvas for plotting
- `geom_`*() creates actual graphical layers
 - `geom_point()`
 - `geom_boxplot()`
- `aes()` defines an 'aesthetic' either globally or by layer<br/>

**Layering**<br/>
Layer 1 is `ggplot()` and layer 2 is + `geom_point()`<br/>
`ggplot(diamonds, aes())` + `geom_point()`

### ggplot density plot
Visualizes the denisty of observations that fall under a certain range of values.

Note the location of aes()

```python
# Load the required package
from plotnine import *
# ggplot density plot
density_p = (ggplot(diamonds) 
 + geom_density(aes(x='carat'), fill='blue')) + xlab("Carat") + ylab("Price")
print(density_p)
```

### ggplot scatter plot

Note the location of aes()

```python
# ggplot scatter plot
scatter_p = (ggplot(diamonds, aes(x='carat', y='price')) + geom_point()) + xlab("Carat") + ylab("Price")
print(scatter_p)
```

### Storing a ggplot object
This allows reuse of code and easy modification or added layers.

```python
# Storing a ggplot object for future modification
g = (ggplot(diamonds, aes(x='carat', y='price')))
g1 = g + geom_point(aes(color='color')) + xlab("Carat") + ylab("Price")
print(g1)
```

```python
g = (ggplot(diamonds, aes(x='carat', y='price')))
g2 = g + geom_point(aes(color='clarity')) + xlab("Carat") + ylab("Price")
print(g2)
```

### Separating segments
Interesting insights and features can be found through segmentation and clear separations of these segments.

```python
# Separating segments
# Note: Python implementation of ggplot facet_wrap() might not always render correctly
#       See R script for how plot should render
g3 = g + geom_point(aes(color='color')) + facet_wrap('~color') + xlab("Carat") + ylab("Price")
print(g3)
```

```python
# Note: Python implementation of ggplot facet_wrap() might not always render correctly
#       See R script for how plot should render
g4 = g + geom_point(aes(color='clarity')) + facet_wrap('~color') + xlab("Carat") + ylab("Price")
print(g4)
```

### Further segmentation
What could we discover if we added more dimensions or variables and their relationships?

```python
# Further segmentation
# Note: Python implementation of ggplot facet_wrap() might not always render correctly
#       See R script for how plot should render
g5 = g + geom_point(aes(color='color')) + facet_wrap('~cut+clarity') + xlab("Carat") + ylab("Price")
print(g5)
```

```python
# Separating out segmentations
# Note: Python implementation of ggplot facet_grid() might not always render correctly
#       See R script for how plot should render
g6 = g + geom_point(aes(color='color')) + facet_grid('cut~clarity') + xlab("Carat") + ylab("Price")
print(g6)
```

## The Titanic dataset

#### Dataset description
**Titanic:** This data set provides information on the fate of passengers on the fatal maiden voyage of the ocean liner "Titanic", summarized according to economic status (class), sex, age, survival and other variables. __[Click here to read more](https://www.rdocumentation.org/packages/titanic/versions/0.1.0)__.

```python
# Read and preview the data
titanic = pd.read_csv("Datasets/titanic.csv")
print(titanic.head())
```

### Check data types

```python
print(titanic.dtypes)
```

### Casting Survived as factor
Treat Survived [0,1] as a categorical variable with factor levels dead ('0') and survived ('1').

```python
# Casting Survived as factor/category levels
titanic['Survived'] = pd.Categorical(titanic.Survived)
```

### Renaming factor levels
Create more meaningful categories or labels for Survived and Embarked.

```python
# Renaming factor levels
print(titanic['Survived'].unique())
titanic['Survived'] = titanic['Survived'].cat.rename_categories(['Dead','Survived'])
titanic['Embarked'] = titanic.Embarked.fillna('Unknown')
titanic['Embarked'] = pd.Categorical(titanic.Embarked)
print(titanic['Embarked'].unique())
titanic['Embarked'] = titanic['Embarked'].cat.rename_categories(['Southampton', 'Cherbourg', 'Queenstown', 'Unknown'])
print(titanic.dtypes)
```

### Check class distribution
What are the portions of Survived and Dead people? Are they fairly balanced?

```python
# Checking class distribution using pie chart
plt.figure(figsize=(12,8))
titanic['Survived'].value_counts().plot(kind='pie')
plt.show()
```

### Is Sex a good predictor?
Plot Dead versus Survived, and store the plot for future modification. Use the object and add a bar plot. The bars for Dead and Survived are segmented by Sex.

```python
# Is Sex a good predictor?
sex_survived_plt = (ggplot(titanic, aes(x='Sex', fill='Survived')))
sex_survived_plt = sex_survived_plt + geom_bar()
print(sex_survived_plt)
```

### Faceting by Sex

We can use `facet_wrap()` to stitch two plots together. So `facet_wrap()` by Sex stitches together the plot for female and the plot for male.

```python
# Faceting by Sex
# Note: Python implementation of ggplot facet_wrap() might not always render correctly
#       See R script for how plot should render
survived_sex_plt = (ggplot(titanic, aes(x='Survived', fill='Sex')))
survived_sex_plt = survived_sex_plt + geom_bar() + facet_wrap('~Sex')
print(survived_sex_plt)
```

### Faceting by Embarked
`facet_wrap()` by Embarked stitches together plots for Cherbourg, Southampton, Queenstown, and Unknown (missing value).

```python
# Faceting by Embarked
# Note: Python implementation of ggplot facet_wrap() might not always render correctly
#       See R script for how plot should render
survived_sex_plt = survived_sex_plt + geom_bar() + facet_wrap('~ Embarked')
print(survived_sex_plt)
```

### Faceting by more than 1 variable
`facet_wrap()` can take only a single variable. So `facet_grid()` is used to facet by multiple variables simultaneously. This is a simultaneous facet by Sex and Embarked.

```python
# Faceting by more than 1 variable
# Note: Python implementation of ggplot facet_grid() might not always render correctly
#       See R script for how plot should render
survived_sex_plt = survived_sex_plt + geom_bar() + facet_grid('Sex ~ Embarked')
print(survived_sex_plt)
```

### Is Age a good predictor?
What are the age ranges and do we see a big difference in age for survivors versus non survivors?

```python
# Is Age a good predictor?
print(titanic['Age'].describe())
titanic_dead = titanic['Survived']=='Dead'
print(titanic['Survived'].head())
print(titanic_dead.head())
print(titanic.where(titanic_dead, inplace=False)['Age'].describe())
titanic_survived = titanic['Survived']=='Survived'
print(titanic.where(titanic_survived, inplace=False)['Age'].describe())
```

### Age distribution by gender
Do we see anything interesting for age between genders?

```python
# NOTE: Please ignore the warning message
# Age distribution by gender
age_sex_plt = (ggplot(titanic, aes(x='Age', fill='Sex')) + geom_histogram())
age_sex_plt = age_sex_plt + labs(x='Distribution of Age', y='Frequency Bucket', 
                   title='Distribution of Passenger Ages on Titanic')
print(age_sex_plt)
```

### Age on survivability
Does Age have an impact on survivability?

```python
# NOTE: Please ignore the warning message
# Age on survivability
survived_age_plt = (ggplot(titanic, aes(x='Survived', y='Age')))
survived_age_plt1 = survived_age_plt + geom_boxplot()
print(survived_age_plt1)
```

### Further exploring Age on survivability
What if plotted Age on surivability a different way? Would we see anything interesting?

```python
# NOTE: Please ignore the warning message
# Further exploring Age on survivability by gender
age_survived_plt = (ggplot(titanic, aes(x='Age', color='Survived')))
age_survived_plt1 = age_survived_plt + geom_density() + ylab("Density")
print(age_survived_plt1)
```

### Segmenting Age on survivability by gender and location
What if we added Sex and Embarked? Would we see anything interesting in these relationships?

```python
# NOTE: Please ignore the warning message
# Segmenting Age on survivability by gender and location
# Note: Python implementation of ggplot facet_grid() might not always render correctly
#       See R script for how plot should render
age_survived_plt2 = age_survived_plt + geom_density() + facet_grid('Sex~Embarked') + ylab("Density")
print(age_survived_plt2)
```

```python
# NOTE: Please ignore the warning message
# Boxplot of this segmentation
# Note: Python implementation of ggplot facet_grid() might not always render correctly
#       See R script for how plot should render
survived_age_plt2 = survived_age_plt + geom_boxplot() + facet_grid('Embarked~Sex')
print(survived_age_plt2)
```
